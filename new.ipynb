```python
# ── GCN Node Activation Visualizer for current/new checkpoints ───────────────
# Shows:
#   left  = temporal graph nodes colored by PCA-1 of conv activations
#   right = raw accelerometer X/Y/Z time series
# Notes:
#   - Works with the current Stage1/Stage2 checkpoint format
#   - Uses the active dual-branch aligner: hip_encoder / wrist_encoder
#   - build_imu_features() now returns 10 channels, not 6
# ──────────────────────────────────────────────────────────────────────────────

import sys, os
sys.path.insert(0, "/home/qsw26/smartfall/gait_loss/ldm_with_loss")
os.environ["CUDA_VISIBLE_DEVICES"] = "1"

import torch
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import networkx as nx
from sklearn.decomposition import PCA

from diffusion_model.dataset import create_dataset
from diffusion_model.model import Stage1Model, Stage2Model
from diffusion_model.model_loader import load_checkpoint, infer_graph_ops_from_checkpoint
from diffusion_model.sensor_model import build_imu_features
from diffusion_model.gait_metrics import DEFAULT_GAIT_METRICS_DIM
from diffusion_model.util import DEFAULT_TIMESTEPS

# ── CONFIG ────────────────────────────────────────────────────────────────────
STAGE1_CKPT = "/home/qsw26/smartfall/gait_loss/ldm_with_loss/new_checkpoints/stage1_gcnn_changed_edits_withreport_april3/stage1_best.pt"
STAGE2_CKPT = "/home/qsw26/smartfall/gait_loss/ldm_with_loss/new_checkpoints/stage2_gcnn_changed_edits_withreport_april3/stage2_best.pt"

SKEL_FOLDER  = "/home/qsw26/smartfall/gait_loss/filtered_skeleton_data"
HIP_FOLDER   = "/home/qsw26/smartfall/gait_loss/SmartFallMM-Dataset/young/accelerometer/phone"
WRIST_FOLDER = "/home/qsw26/smartfall/gait_loss/SmartFallMM-Dataset/young/accelerometer/watch"

SAMPLE_IDX = 67
GCN_LAYER  = "conv3"       # "conv1" | "conv2" | "conv3"
BRANCH     = "wrist"       # "hip" | "wrist"
DEVICE     = "cpu"         # "cpu" | "cuda"
WINDOW     = 90
STRIDE     = 45
NUM_JOINTS = 32
NUM_CLASSES = 14
LATENT_DIM = 256
D_SHARED   = 64
IMU_GRAPH  = "multiscale"  # must match how Stage2 was trained

ACTIVITY_NAMES = {
    0:"Walking",1:"Running",2:"Shuffling",3:"Stairs (up)",4:"Stairs (down)",
    5:"Standing",6:"Sitting",7:"Lying",8:"Sit-to-stand",9:"Stand-to-sit",
    10:"Bending",11:"Reaching",12:"Driving",13:"Cycling",
}

# ── Load models ───────────────────────────────────────────────────────────────
device = torch.device(DEVICE)

encoder_graph_op, skeleton_graph_op = infer_graph_ops_from_checkpoint(STAGE1_CKPT)

stage1 = Stage1Model(
    latent_dim=128,
    num_joints=NUM_JOINTS,
    timesteps=DEFAULT_TIMESTEPS,
    gait_metrics_dim=DEFAULT_GAIT_METRICS_DIM,
    use_gait_conditioning=False,
    num_classes=NUM_CLASSES,
    encoder_type=encoder_graph_op,
    skeleton_graph_op=skeleton_graph_op,
).to(device)

load_checkpoint(STAGE1_CKPT, stage1, strict=False)
stage1.eval()

stage2 = Stage2Model(
    encoder=stage1.encoder,
    latent_dim=128,
    num_joints=32,
    gait_metrics_dim=9,
    num_classes=14,
    imu_graph_type=IMU_GRAPH,
    d_shared=D_SHARED,
).to(device)

ckpt = torch.load(STAGE2_CKPT, map_location="cpu")
state = ckpt["state_dict"] if "state_dict" in ckpt else ckpt
aligner_state = {k[len("aligner."):]: v for k, v in state.items() if k.startswith("aligner.")}
stage2.aligner.load_state_dict(aligner_state, strict=True)

stage2.eval()


# ── Load one sample ───────────────────────────────────────────────────────────
ds = create_dataset(
    dataset_path=None,
    skeleton_folder=SKEL_FOLDER,
    hip_folder=HIP_FOLDER,
    wrist_folder=WRIST_FOLDER,
    window=WINDOW,
    joints=NUM_JOINTS,
    num_classes=NUM_CLASSES,
    stride=STRIDE,
    normalize_sensors=True,
    gait_cache_dir="outputs/gait_cache",
)

sample = ds[SAMPLE_IDX]
label = int(sample["label"].item())
a_hip = sample["A_hip"].unsqueeze(0).to(device)       # [1, T, 3]
a_wrist = sample["A_wrist"].unsqueeze(0).to(device)   # [1, T, 3]
T = a_hip.shape[1]

# ── Hook into the chosen Stage2 sensor branch ────────────────────────────────
encoder = stage2.aligner.hip_encoder if BRANCH == "hip" else stage2.aligner.wrist_encoder
gcn_layer = getattr(encoder, GCN_LAYER)   # conv1 / conv2 / conv3

captured = {}

def _hook(module, inp, out):
    # out shape: [B*T, hidden_dim]
    captured["act"] = out.detach().cpu()

handle = gcn_layer.register_forward_hook(_hook)

with torch.no_grad():
    stream = a_hip if BRANCH == "hip" else a_wrist
    feat = build_imu_features(stream)   # [1, T, 10] in current code
    _ = encoder(feat)

handle.remove()

# ── PCA reduce hidden activations to one scalar per timestep ─────────────────
act = captured["act"].numpy()           # [T, hidden_dim]
pca = PCA(n_components=1)
pca_vals = pca.fit_transform(act).squeeze()
pca_norm = (pca_vals - pca_vals.min()) / (np.ptp(pca_vals) + 1e-8)

# ── Build simple temporal chain graph for layout ─────────────────────────────
G = nx.Graph()
G.add_nodes_from(range(T))
for i in range(T - 1):
    G.add_edge(i, i + 1)

pos = nx.spring_layout(G, seed=1, k=0.4, iterations=80)

# ── Plot ──────────────────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Left: graph nodes colored by PCA-reduced activation
ax = axes[0]
nx.draw_networkx_edges(G, pos, ax=ax, edge_color="#cccccc", width=0.8, alpha=0.6)
nx.draw_networkx_nodes(
    G, pos, ax=ax,
    node_color=pca_norm,
    cmap="viridis",
    node_size=220,
    alpha=0.92,
)
nx.draw_networkx_labels(
    G, pos, ax=ax,
    labels={i: str(i) for i in range(T)},
    font_size=5,
    font_color="white",
)
ax.set_title(
    f"{BRANCH.title()} {GCN_LAYER} Activations (PCA-1)\n"
    f"Label: {ACTIVITY_NAMES.get(label, f'A{label+1:02d}')}"
)
ax.axis("off")
cbar = fig.colorbar(
    plt.cm.ScalarMappable(cmap="viridis", norm=plt.Normalize(0, 1)),
    ax=ax,
    fraction=0.035,
    pad=0.02,
)
cbar.set_label("PCA Component 1")

# Right: raw accel stream
ax2 = axes[1]
accel_np = (a_hip if BRANCH == "hip" else a_wrist).squeeze(0).cpu().numpy()  # [T, 3]
frames = np.arange(T)
ax2.plot(frames, accel_np[:, 0], color="red", linewidth=1.8, label="X-axis")
ax2.plot(frames, accel_np[:, 1], color="blue", linewidth=1.8, label="Y-axis")
ax2.plot(frames, accel_np[:, 2], color="green", linewidth=1.8, label="Z-axis")
ax2.set_title(
    f"Raw Accelerometer ({BRANCH.title()})\n"
    f"Label: {ACTIVITY_NAMES.get(label, f'A{label+1:02d}')}"
)
ax2.set_xlabel("Time Step")
ax2.set_ylabel("Acceleration")
ax2.legend(loc="upper right")
ax2.grid(True, alpha=0.25)

fig.tight_layout()

out_path = f"outputs/gcn_node_activations_sample{SAMPLE_IDX}_{BRANCH}_{GCN_LAYER}.png"
plt.savefig(out_path, dpi=150, bbox_inches="tight")
plt.show()

print(f"Saved: {out_path}")
print(f"PCA explained variance: {pca.explained_variance_ratio_[0] * 100:.1f}%")
print(f"Label: {label} -> {ACTIVITY_NAMES.get(label, f'A{label+1:02d}')}")
print(f"Activation shape: {act.shape}")
```

```python
# ── Stage2 IMU Activation Visualizer Suite ───────────────────────────────────
# Includes:
#   1. plot_branch_activation(...)   -> hip/wrist branch conv1/2/3 activations
#   2. plot_fused_tokens(...)        -> fused sensor_tokens after aligner fusion
#   3. export_many_samples(...)      -> save many figures in a loop
# ──────────────────────────────────────────────────────────────────────────────

import sys, os
sys.path.insert(0, "/home/qsw26/smartfall/gait_loss/ldm_with_loss")
os.environ["CUDA_VISIBLE_DEVICES"] = "1"

import torch
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from sklearn.decomposition import PCA

from diffusion_model.dataset import create_dataset
from diffusion_model.model import Stage1Model, Stage2Model
from diffusion_model.model_loader import load_checkpoint, infer_graph_ops_from_checkpoint
from diffusion_model.sensor_model import build_imu_features
from diffusion_model.gait_metrics import DEFAULT_GAIT_METRICS_DIM
from diffusion_model.util import DEFAULT_TIMESTEPS

# ── CONFIG ────────────────────────────────────────────────────────────────────
STAGE1_CKPT = "/home/qsw26/smartfall/gait_loss/ldm_with_loss/new_checkpoints/stage1_gcnn_changed_edits_withreport_april3/stage1_best.pt"
STAGE2_CKPT = "/home/qsw26/smartfall/gait_loss/ldm_with_loss/new_checkpoints/stage2_gcnn_changed_edits_withreport_april3/stage2_best.pt"

SKEL_FOLDER  = "/home/qsw26/smartfall/gait_loss/filtered_skeleton_data"
HIP_FOLDER   = "/home/qsw26/smartfall/gait_loss/SmartFallMM-Dataset/young/accelerometer/phone"
WRIST_FOLDER = "/home/qsw26/smartfall/gait_loss/SmartFallMM-Dataset/young/accelerometer/watch"

DEVICE = "cpu"
WINDOW = 90
STRIDE = 45
NUM_JOINTS = 32
NUM_CLASSES = 14
LATENT_DIM = 256
D_SHARED = 64
IMU_GRAPH = "multiscale"
OUT_DIR = "outputs/stage2_activation_viz"

ACTIVITY_NAMES = {8:"walking"}

os.makedirs(OUT_DIR, exist_ok=True)

# ── Load models ───────────────────────────────────────────────────────────────
device = torch.device(DEVICE)
encoder_graph_op, skeleton_graph_op = infer_graph_ops_from_checkpoint(STAGE1_CKPT)

stage2 = Stage2Model(
    encoder=stage1.encoder,
    latent_dim=128,
    num_joints=32,
    gait_metrics_dim=9,
    num_classes=14,
    imu_graph_type=IMU_GRAPH,
    d_shared=D_SHARED,
).to(device)

ckpt = torch.load(STAGE2_CKPT, map_location="cpu")
state = ckpt["state_dict"] if "state_dict" in ckpt else ckpt
aligner_state = {k[len("aligner."):]: v for k, v in state.items() if k.startswith("aligner.")}
stage2.aligner.load_state_dict(aligner_state, strict=True)

stage2.eval()


stage2 = Stage2Model(
    encoder=stage1.encoder,
    latent_dim=128,
    num_joints=32,
    gait_metrics_dim=9,
    num_classes=14,
    imu_graph_type=IMU_GRAPH,
    d_shared=D_SHARED,
).to(device)

ckpt = torch.load(STAGE2_CKPT, map_location="cpu")
state = ckpt["state_dict"] if "state_dict" in ckpt else ckpt
aligner_state = {k[len("aligner."):]: v for k, v in state.items() if k.startswith("aligner.")}
stage2.aligner.load_state_dict(aligner_state, strict=True)

stage2.eval()


# ── Dataset ───────────────────────────────────────────────────────────────────
ds = create_dataset(
    dataset_path=None,
    skeleton_folder=SKEL_FOLDER,
    hip_folder=HIP_FOLDER,
    wrist_folder=WRIST_FOLDER,
    window=WINDOW,
    joints=NUM_JOINTS,
    num_classes=NUM_CLASSES,
    stride=STRIDE,
    normalize_sensors=True,
    gait_cache_dir="outputs/gait_cache",
)

def get_sample(sample_idx):
    sample = ds[sample_idx]
    return {
        "label": int(sample["label"].item()),
        "a_hip": sample["A_hip"].unsqueeze(0).to(device),
        "a_wrist": sample["A_wrist"].unsqueeze(0).to(device),
    }

def build_chain_graph(T):
    G = nx.Graph()
    G.add_nodes_from(range(T))
    for i in range(T - 1):
        G.add_edge(i, i + 1)
    pos = nx.spring_layout(G, seed=1, k=0.4, iterations=80)
    return G, pos

def reduce_to_scalar(features_2d):
    pca = PCA(n_components=1)
    vals = pca.fit_transform(features_2d).squeeze()
    vals = np.asarray(vals, dtype=np.float32)
    vals_norm = (vals - vals.min()) / (np.ptp(vals) + 1e-8)
    return vals_norm, pca.explained_variance_ratio_[0]

def plot_graph_and_signal(node_values, accel_np, title_left, title_right, out_path, support_text=None):
    T = len(node_values)
    G, pos = build_chain_graph(T)

    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    ax = axes[0]
    nx.draw_networkx_edges(G, pos, ax=ax, edge_color="#cccccc", width=0.8, alpha=0.6)
    nx.draw_networkx_nodes(
        G, pos, ax=ax,
        node_color=node_values,
        cmap="viridis",
        node_size=220,
        alpha=0.92,
    )
    nx.draw_networkx_labels(
        G, pos, ax=ax,
        labels={i: str(i) for i in range(T)},
        font_size=5,
        font_color="white",
    )
    ax.set_title(title_left)
    ax.axis("off")
    cbar = fig.colorbar(
        plt.cm.ScalarMappable(cmap="viridis", norm=plt.Normalize(0, 1)),
        ax=ax, fraction=0.035, pad=0.02
    )
    cbar.set_label("PCA Component 1")

    ax2 = axes[1]
    frames = np.arange(T)
    ax2.plot(frames, accel_np[:, 0], color="red", linewidth=1.8, label="X-axis")
    ax2.plot(frames, accel_np[:, 1], color="blue", linewidth=1.8, label="Y-axis")
    ax2.plot(frames, accel_np[:, 2], color="green", linewidth=1.8, label="Z-axis")
    ax2.set_title(title_right)
    ax2.set_xlabel("Time Step")
    ax2.set_ylabel("Acceleration")
    ax2.legend(loc="upper right")
    ax2.grid(True, alpha=0.25)

    if support_text is not None:
        fig.suptitle(support_text, y=1.02, fontsize=11)

    fig.tight_layout()
    plt.savefig(out_path, dpi=150, bbox_inches="tight")
    plt.show()

def plot_branch_activation(sample_idx=17, branch="wrist", gcn_layer="conv3"):
    sample = get_sample(sample_idx)
    label = sample["label"]
    a_hip = sample["a_hip"]
    a_wrist = sample["a_wrist"]

    encoder = stage2.aligner.hip_encoder if branch == "hip" else stage2.aligner.wrist_encoder
    layer = getattr(encoder, gcn_layer)

    captured = {}

    def _hook(module, inp, out):
        captured["act"] = out.detach().cpu()

    handle = layer.register_forward_hook(_hook)

    with torch.no_grad():
        stream = a_hip if branch == "hip" else a_wrist
        feat = build_imu_features(stream)   # [1, T, 10]
        _ = encoder(feat)

    handle.remove()

    act = captured["act"].numpy()          # [T, hidden_dim]
    node_values, evr = reduce_to_scalar(act)
    accel_np = (a_hip if branch == "hip" else a_wrist).squeeze(0).cpu().numpy()

    out_path = os.path.join(OUT_DIR, f"sample{sample_idx}_{branch}_{gcn_layer}.png")
    plot_graph_and_signal(
        node_values=node_values,
        accel_np=accel_np,
        title_left=f"{branch.title()} {gcn_layer} Activations (PCA-1)\nLabel: {ACTIVITY_NAMES.get(label, f'A{label+1:02d}')}",
        title_right=f"Raw Accelerometer ({branch.title()})\nLabel: {ACTIVITY_NAMES.get(label, f'A{label+1:02d}')}",
        out_path=out_path,
        support_text=f"Sample {sample_idx} | PCA explained variance: {evr*100:.1f}%",
    )
    print(f"Saved: {out_path}")
    print(f"Label: {label} -> {ACTIVITY_NAMES.get(label, f'A{label+1:02d}')}")
    print(f"Activation shape: {act.shape}")
    print(f"PCA explained variance: {evr*100:.1f}%")

def plot_fused_tokens(sample_idx=17):
    sample = get_sample(sample_idx)
    label = sample["label"]
    a_hip = sample["a_hip"]
    a_wrist = sample["a_wrist"]

    with torch.no_grad():
        sensor_tokens, h_global = stage2.aligner(a_hip, a_wrist)   # [1, T, D], [1, D]

    tokens = sensor_tokens.squeeze(0).cpu().numpy()   # [T, D]
    node_values, evr = reduce_to_scalar(tokens)

    accel_np = 0.5 * (
        a_hip.squeeze(0).cpu().numpy() +
        a_wrist.squeeze(0).cpu().numpy()
    )

    out_path = os.path.join(OUT_DIR, f"sample{sample_idx}_fused_tokens.png")
    plot_graph_and_signal(
        node_values=node_values,
        accel_np=accel_np,
        title_left=f"Fused sensor_tokens (PCA-1)\nLabel: {ACTIVITY_NAMES.get(label, f'A{label+1:02d}')}",
        title_right=f"Mean Raw Accelerometer (Hip + Wrist)/2\nLabel: {ACTIVITY_NAMES.get(label, f'A{label+1:02d}')}",
        out_path=out_path,
        support_text=f"Sample {sample_idx} | PCA explained variance: {evr*100:.1f}%",
    )
    print(f"Saved: {out_path}")
    print(f"Label: {label} -> {ACTIVITY_NAMES.get(label, f'A{label+1:02d}')}")
    print(f"sensor_tokens shape: {tokens.shape}")
    print(f"h_global shape: {tuple(h_global.shape)}")
    print(f"PCA explained variance: {evr*100:.1f}%")

def export_many_samples(sample_ids, branch="wrist", gcn_layer="conv3", also_fused=True):
    for sample_idx in sample_ids:
        try:
            plot_branch_activation(sample_idx=sample_idx, branch=branch, gcn_layer=gcn_layer)
            if also_fused:
                plot_fused_tokens(sample_idx=sample_idx)
        except Exception as e:
            print(f"Failed sample {sample_idx}: {e}")

# ── Examples ──────────────────────────────────────────────────────────────────
# Single branch-layer view
plot_branch_activation(sample_idx=72, branch="wrist", gcn_layer="conv3")

# Single fused-token view
plot_fused_tokens(sample_idx=72)

# Batch export
# export_many_samples(sample_ids=[0, 5, 17, 25, 40], branch="hip", gcn_layer="conv2", also_fused=True)
```
