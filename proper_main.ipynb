## Code to visualize the before and after the encoder decoder

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

from diffusion_model.dataset import create_dataloader
from diffusion_model.model import Stage1Model
from diffusion_model.model_loader import load_checkpoint
from diffusion_model.util import get_skeleton_edges, DEFAULT_JOINTS, DEFAULT_LATENT_DIM, DEFAULT_TIMESTEPS
from diffusion_model.gait_metrics import DEFAULT_GAIT_METRICS_DIM

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

stage1_ckpt = "outputs/retrain_meeting_march_13/checkpoints/stage1_best.pt"

loader = create_dataloader(
    dataset_path=None,
    batch_size=1,
    window=90,
    joints=32,
    num_classes=14,
    skeleton_folder="/home/qsw26/smartfall/gait_loss/filtered_skeleton_data",
    hip_folder="/home/qsw26/smartfall/gait_loss/SmartFallMM-Dataset/young/accelerometer/meta_hip",
    wrist_folder="/home/qsw26/smartfall/gait_loss/SmartFallMM-Dataset/young/accelerometer/meta_wrist",
    stride=45,
    normalize_sensors=True,
    shuffle=True,
    gait_cache_dir="outputs/retrain_meeting/gait_cache_v2_9metrics",
)

model = Stage1Model(
    latent_dim=DEFAULT_LATENT_DIM,
    num_joints=DEFAULT_JOINTS,
    timesteps=500,
    gait_metrics_dim=DEFAULT_GAIT_METRICS_DIM,
).to(device)
load_checkpoint(stage1_ckpt, model, strict=True)
model.eval()

batch = next(iter(loader))
x = batch["skeleton"].to(device)         # [1, T, J, 3]
gait_metrics = batch["gait_metrics"].to(device)

with torch.no_grad():
    z0 = model.encoder(x, gait_metrics=gait_metrics)   # [1, T, J, D]
    x_recon = model.decoder(z0)                        # [1, T, J, 3]

x_np = x[0].cpu().numpy()
z_np = z0[0].cpu().numpy()
x_recon_np = x_recon[0].cpu().numpy()

latent_norm = np.linalg.norm(z_np, axis=-1)  # [T, J]

edges = get_skeleton_edges()
mid = x_np.shape[0] // 2

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Before encoder: raw skeleton
frame = x_np[mid]
for i, j in edges:
    if i < frame.shape[0] and j < frame.shape[0]:
        axes[0].plot([frame[i, 0], frame[j, 0]], [frame[i, 1], frame[j, 1]], color="black", linewidth=1.5)
axes[0].scatter(frame[:, 0], frame[:, 1], s=20, c="royalblue")
axes[0].invert_yaxis()
axes[0].set_aspect("equal")
axes[0].set_title("Before Encoder: Raw Skeleton")
axes[0].axis("off")

# 2. After encoder: latent activation map
im = axes[1].imshow(latent_norm.T, aspect="auto", cmap="magma", interpolation="nearest")
axes[1].set_title("After Encoder: Latent Norm Map")
axes[1].set_xlabel("Frame")
axes[1].set_ylabel("Joint")
plt.colorbar(im, ax=axes[1], fraction=0.046, pad=0.04)

# 3. Decoder reconstruction from encoded latent
frame_hat = x_recon_np[mid]
for i, j in edges:
    if i < frame_hat.shape[0] and j < frame_hat.shape[0]:
        axes[2].plot([frame_hat[i, 0], frame_hat[j, 0]], [frame_hat[i, 1], frame_hat[j, 1]], color="black", linewidth=1.5)
axes[2].scatter(frame_hat[:, 0], frame_hat[:, 1], s=20, c="crimson")
axes[2].invert_yaxis()
axes[2].set_aspect("equal")
axes[2].set_title("Decoded From Encoder Latent")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

from diffusion_model.dataset import create_dataloader
from diffusion_model.model import Stage1Model, Stage2Model
from diffusion_model.model_loader import load_checkpoint
from diffusion_model.util import DEFAULT_JOINTS, DEFAULT_LATENT_DIM
from diffusion_model.gait_metrics import DEFAULT_GAIT_METRICS_DIM

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

try:
    import umap
except ImportError:
    umap = None

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

stage1_ckpt = "outputs/retrain_meeting_march_13/checkpoints/stage1_best.pt"
stage2_ckpt = "outputs/retrain_meeting_march_13/checkpoints/stage2_best.pt"

loader = create_dataloader(
    dataset_path=None,
    batch_size=16,
    window=90,
    joints=32,
    num_classes=14,
    skeleton_folder="/home/qsw26/smartfall/gait_loss/filtered_skeleton_data",
    hip_folder="/home/qsw26/smartfall/gait_loss/SmartFallMM-Dataset/young/accelerometer/meta_hip",
    wrist_folder="/home/qsw26/smartfall/gait_loss/SmartFallMM-Dataset/young/accelerometer/meta_wrist",
    stride=45,
    normalize_sensors=True,
    shuffle=True,
    gait_cache_dir="outputs/retrain_meeting/gait_cache_v2_9metrics",
)

stage1 = Stage1Model(
    latent_dim=DEFAULT_LATENT_DIM,
    num_joints=DEFAULT_JOINTS,
    timesteps=500,
    gait_metrics_dim=DEFAULT_GAIT_METRICS_DIM,
).to(device)
load_checkpoint(stage1_ckpt, stage1, strict=True)
stage1.eval()

stage2 = Stage2Model(
    encoder=stage1.encoder,
    latent_dim=DEFAULT_LATENT_DIM,
    num_joints=DEFAULT_JOINTS,
    gait_metrics_dim=DEFAULT_GAIT_METRICS_DIM,
).to(device)
load_checkpoint(stage2_ckpt, stage2, strict=True)
stage2.eval()

latent_features = []
sensor_features = []
labels = []

max_batches = 8

with torch.no_grad():
    for batch_idx, batch in enumerate(loader):
        if batch_idx >= max_batches:
            break

        x = batch["skeleton"].to(device)
        a_hip = batch["A_hip"].to(device)
        a_wrist = batch["A_wrist"].to(device)
        gait_metrics = batch["gait_metrics"].to(device)
        y = batch["label"].cpu().numpy()

        z0 = stage1.encoder(x, gait_metrics=gait_metrics)          # [B, T, J, D]
        _, h_global = stage2.aligner(a_hip, a_wrist, gait_metrics=gait_metrics)  # [B, D]

        latent_pool = z0.mean(dim=(1, 2)).cpu().numpy()           # [B, D]
        sensor_pool = h_global.cpu().numpy()                      # [B, D]

        latent_features.append(latent_pool)
        sensor_features.append(sensor_pool)
        labels.append(y)

latent_arr = np.concatenate(latent_features, axis=0)
sensor_arr = np.concatenate(sensor_features, axis=0)
label_arr = np.concatenate(labels, axis=0)

def project_features(features, method="pca"):
    if method == "pca":
        reducer = PCA(n_components=2, random_state=42)
        return reducer.fit_transform(features)
    elif method == "tsne":
        perplexity = min(30, max(5, features.shape[0] // 4))
        reducer = TSNE(n_components=2, init="pca", learning_rate="auto", perplexity=perplexity, random_state=42)
        return reducer.fit_transform(features)
    elif method == "umap":
        if umap is None:
            raise ImportError("umap-learn is not installed. Run: pip install umap-learn")
        reducer = umap.UMAP(n_components=2, n_neighbors=min(15, features.shape[0] - 1), min_dist=0.15, random_state=42)
        return reducer.fit_transform(features)
    else:
        raise ValueError(f"Unknown method: {method}")

methods = ["pca", "tsne", "umap"]
titles = ["PCA", "t-SNE", "UMAP"]

fig, axes = plt.subplots(2, 3, figsize=(18, 10))

for col, (method, title) in enumerate(zip(methods, titles)):
    latent_proj = project_features(latent_arr, method=method)
    sensor_proj = project_features(sensor_arr, method=method)

    sc1 = axes[0, col].scatter(
        latent_proj[:, 0], latent_proj[:, 1],
        c=label_arr, cmap="tab20", s=30, alpha=0.85
    )
    axes[0, col].set_title(f"Stage2 Skeleton Latent {title}")
    axes[0, col].set_xlabel("Dim 1")
    axes[0, col].set_ylabel("Dim 2")
    axes[0, col].grid(alpha=0.2)

    sc2 = axes[1, col].scatter(
        sensor_proj[:, 0], sensor_proj[:, 1],
        c=label_arr, cmap="tab20", s=30, alpha=0.85
    )
    axes[1, col].set_title(f"Stage2 Sensor Embedding {title}")
    axes[1, col].set_xlabel("Dim 1")
    axes[1, col].set_ylabel("Dim 2")
    axes[1, col].grid(alpha=0.2)

cbar = fig.colorbar(sc2, ax=axes, shrink=0.8)
cbar.set_label("Class Label")

plt.tight_layout()
plt.show()
```
