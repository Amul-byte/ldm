```python
# ── 4-Panel UMAP: Raw | Stage-1 | Stage-2 IMU | Stage-3 Generated ────────────
import sys, os
sys.path.insert(0, "/home/qsw26/smartfall/gait_loss/ldm_with_loss")
import sklearn
tsne_kwargs = {"n_components": 2, "random_state": 42, "perplexity": 30, "n_jobs": -1}
tsne_kwargs["max_iter" if tuple(int(x) for x in sklearn.__version__.split(".")[:2]) >= (1, 4) else "n_iter"] = 1000

import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import umap
from sklearn.decomposition import PCA
from torch.utils.data import DataLoader, random_split

from diffusion_model.dataset import create_dataset
from diffusion_model.model import Stage1Model, Stage2Model, Stage3Model
from diffusion_model.model_loader import load_checkpoint

# ── CONFIG ───────────────────────────────────────────────────────────────────
STAGE1_CKPT  = "new_checkpoints/gcnn_stage1/stage1_best.pt"
STAGE2_CKPT  = "new_checkpoints/stage1_gcnn_mseonly/stage2_best.pt"
STAGE3_CKPT  = "new_checkpoints/stage3_gcnn/stage3_best.pt"
SKEL_FOLDER  = "/home/qsw26/smartfall/gait_loss/filtered_skeleton_data"
HIP_FOLDER   = "/home/qsw26/smartfall/gait_loss/SmartFallMM-Dataset/young/accelerometer/phone"
WRIST_FOLDER = "/home/qsw26/smartfall/gait_loss/SmartFallMM-Dataset/young/accelerometer/watch"
ENCODER_TYPE = "gcn"
IMU_GRAPH    = "multiscale"
VAL_FRAC     = 0.2
BATCH_SIZE   = 16
DDIM_STEPS   = 20
MAX_S3_SAMPLES = 400      # Stage-3 subset (DDIM is slow); panels 1-3 use full dataset
PCA_COMPONENTS = 64
PROBE_EPOCHS   = 50
NUM_CLASSES    = 14
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
OUT_PATH = "outputs/umap_4stage_comparison.png"

# ── Load models ───────────────────────────────────────────────────────────────
device = torch.device(DEVICE)
print("Loading Stage-1 …")
stage1 = Stage1Model(latent_dim=256, num_joints=32, timesteps=500,
                     gait_metrics_dim=9, encoder_type=ENCODER_TYPE).to(device)
load_checkpoint(STAGE1_CKPT, stage1, strict=False)
stage1.eval()
for p in stage1.parameters(): p.requires_grad_(False)

print("Loading Stage-2 …")
stage2 = Stage2Model(encoder=stage1.encoder, latent_dim=256, num_joints=32,
                     gait_metrics_dim=9, num_classes=NUM_CLASSES,
                     imu_graph_type=IMU_GRAPH).to(device)
load_checkpoint(STAGE2_CKPT, stage2, strict=False)
stage2.eval()
for p in stage2.parameters(): p.requires_grad_(False)

print("Loading Stage-3 …")
stage3 = Stage3Model(
    encoder=stage1.encoder, decoder=stage1.decoder, denoiser=stage1.denoiser,
    latent_dim=256, num_joints=32, num_classes=NUM_CLASSES,
    timesteps=500, gait_metrics_dim=9,
).to(device)
load_checkpoint(STAGE3_CKPT, stage3, strict=False)
stage3.eval()
for p in stage3.parameters(): p.requires_grad_(False)

# ── Dataset ───────────────────────────────────────────────────────────────────
print("Building dataset …")
ds = create_dataset(None,
    skeleton_folder=SKEL_FOLDER, hip_folder=HIP_FOLDER, wrist_folder=WRIST_FOLDER,
    window=90, joints=32, num_classes=NUM_CLASSES, stride=45,
    normalize_sensors=True, gait_cache_dir="outputs/gait_cache",
)
n_val = int(len(ds) * VAL_FRAC)
train_ds, val_ds = random_split(ds, [len(ds)-n_val, n_val],
                                generator=torch.Generator().manual_seed(42))
print(f"  train={len(train_ds)}  val={len(val_ds)}")

# ── Encode all 4 representations ─────────────────────────────────────────────
@torch.no_grad()
def encode_all(loader, collect_s3=False, max_s3=MAX_S3_SAMPLES):
    raw_list, skel_list, imu_list, y_list = [], [], [], []
    s3_list, y_s3_list = [], []   # ← separate label list for Stage-3 (subset)
    s3_count = 0

    for batch in loader:
        x            = batch["skeleton"].to(device)
        a_hip        = batch["A_hip"].to(device)
        a_wrist      = batch["A_wrist"].to(device)
        gait_metrics = batch["gait_metrics"].to(device)
        y_batch      = batch["label"].numpy()
        B = x.shape[0]

        # Panels 1-3: always collected (full dataset)
        raw_list.append(x.cpu().numpy().reshape(B, -1))
        z0_real = stage1.encoder(x, gait_metrics=None)
        skel_list.append(z0_real.cpu().numpy().reshape(B, -1))
        h_tokens, h_global = stage2.aligner(a_hip_stream=a_hip, a_wrist_stream=a_wrist)
        imu_list.append(h_tokens.cpu().numpy().reshape(B, -1))
        y_list.append(y_batch)

        # Panel 4: Stage-3 DDIM — capped at max_s3, track its own labels
        if collect_s3 and s3_count < max_s3:
            z0_gen = stage3.diffusion.p_sample_loop_ddim(
                denoiser=stage3.denoiser,
                shape=torch.Size([B, x.shape[1], x.shape[2], 256]),
                device=device,
                h_tokens=h_tokens,
                h_global=h_global,
                gait_metrics=gait_metrics,
                sample_steps=DDIM_STEPS,
            )
            s3_list.append(z0_gen.cpu().numpy().reshape(B, -1))
            y_s3_list.append(y_batch)   # ← labels only for this subset
            s3_count += B

    return (
        np.concatenate(raw_list),
        np.concatenate(skel_list),
        np.concatenate(imu_list),
        np.concatenate(y_list),
        np.concatenate(s3_list)   if s3_list   else None,
        np.concatenate(y_s3_list) if y_s3_list else None,
    )

print("Encoding train …")
raw_tr, skel_tr, imu_tr, y_tr, s3_tr, y_s3_tr = encode_all(
    DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=0), collect_s3=True)
print("Encoding val …")
raw_va, skel_va, imu_va, y_va, s3_va, y_s3_va = encode_all(
    DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=0), collect_s3=True)
print(f"  Stage-3 train samples: {len(s3_tr)}  val samples: {len(s3_va)}")

# ── PCA ───────────────────────────────────────────────────────────────────────
def pca_compress(tr, va):
    n = min(PCA_COMPONENTS, tr.shape[0], tr.shape[1])
    pca = PCA(n_components=n, random_state=42)
    return pca.fit_transform(tr), pca.transform(va)

print("PCA …")
raw_tr_p,  raw_va_p  = pca_compress(raw_tr,  raw_va)
skel_tr_p, skel_va_p = pca_compress(skel_tr, skel_va)
imu_tr_p,  imu_va_p  = pca_compress(imu_tr,  imu_va)
s3_tr_p,   s3_va_p   = pca_compress(s3_tr,   s3_va)

# ── Linear probes ─────────────────────────────────────────────────────────────
def run_probe(tr, y_tr, va, y_va, name):
    probe = nn.Linear(tr.shape[1], NUM_CLASSES).to(device)
    opt   = torch.optim.Adam(probe.parameters(), lr=1e-3, weight_decay=1e-4)
    xt = torch.tensor(tr, dtype=torch.float32).to(device)
    yt = torch.tensor(y_tr, dtype=torch.long).to(device)
    xv = torch.tensor(va, dtype=torch.float32).to(device)
    yv = torch.tensor(y_va, dtype=torch.long).to(device)
    for _ in range(PROBE_EPOCHS):
        probe.train(); opt.zero_grad()
        nn.CrossEntropyLoss()(probe(xt), yt).backward(); opt.step()
    with torch.no_grad():
        acc = (probe(xv).argmax(1) == yv).float().mean().item()
    print(f"  [{name}] probe acc = {acc:.3f}")
    return acc

print("Linear probes …")
acc_skel = run_probe(skel_tr_p, y_tr,    skel_va_p, y_va,    "Stage-1 real z0")
acc_imu  = run_probe(imu_tr_p,  y_tr,    imu_va_p,  y_va,    "Stage-2 IMU")
acc_s3   = run_probe(s3_tr_p,   y_s3_tr, s3_va_p,   y_s3_va, "Stage-3 generated")  # ← own labels

# ── UMAP ─────────────────────────────────────────────────────────────────────
print("UMAP (takes ~3 min) …")
reducer = umap.UMAP(n_components=2, random_state=42, n_jobs=1)
emb_raw  = reducer.fit_transform(np.concatenate([raw_tr_p,  raw_va_p]));  print("  raw  done")
emb_skel = reducer.fit_transform(np.concatenate([skel_tr_p, skel_va_p])); print("  skel done")
emb_imu  = reducer.fit_transform(np.concatenate([imu_tr_p,  imu_va_p]));  print("  imu  done")
emb_s3   = reducer.fit_transform(np.concatenate([s3_tr_p,   s3_va_p]));   print("  s3   done")
all_y     = np.concatenate([y_tr,    y_va])
all_y_s3  = np.concatenate([y_s3_tr, y_s3_va])

# ── Plot ──────────────────────────────────────────────────────────────────────
cmap = plt.get_cmap("tab20", NUM_CLASSES)
fig, axes = plt.subplots(1, 4, figsize=(30, 7))

def scatter_panel(ax, emb, labels, title, subtitle):
    for c in range(NUM_CLASSES):
        idx = labels == c
        if idx.sum():
            ax.scatter(emb[idx,0], emb[idx,1], s=12, alpha=0.65,
                       color=cmap(c), label=f"A{c+1}")
    ax.legend(markerscale=2, fontsize=6, ncol=2, loc="best")
    ax.set_title(f"{title}\n{subtitle}", fontsize=9)
    ax.set_xlabel("UMAP-1"); ax.set_ylabel("UMAP-2")

scatter_panel(axes[0], emb_raw,  all_y,    "Raw skeleton",              "Full sequence (PCA→UMAP)")
scatter_panel(axes[1], emb_skel, all_y,    "Stage-1 skeleton latents",  f"Real z0  |  probe acc={acc_skel:.3f}")
scatter_panel(axes[2], emb_imu,  all_y,    "Stage-2 IMU latents",       f"h_tokens  |  probe acc={acc_imu:.3f}")
scatter_panel(axes[3], emb_s3,   all_y_s3, "Stage-3 generated latents", f"DDIM z0_gen ({DDIM_STEPS} steps)  |  probe acc={acc_s3:.3f}")

chance = 1.0 / NUM_CLASSES
fig.suptitle(
    f"All-stage representation comparison  (chance={chance:.3f})\n"
    f"Stage-1={acc_skel:.3f}  |  Stage-2 IMU={acc_imu:.3f}  |  Stage-3 generated={acc_s3:.3f}",
    fontsize=12, fontweight="bold",
)
plt.tight_layout()
os.makedirs("outputs", exist_ok=True)
plt.savefig(OUT_PATH, dpi=120, bbox_inches="tight")
plt.show()
print(f"\nSaved → {OUT_PATH}")
```

```python
# ── 4-Panel t-SNE: Raw | Stage-1 | Stage-2 IMU | Stage-3 Generated ────────────
import sys, os
sys.path.insert(0, "/home/qsw26/smartfall/gait_loss/ldm_with_loss")

import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from torch.utils.data import DataLoader, random_split
import sklearn
tsne_kwargs = {"n_components": 2, "random_state": 42, "perplexity": 30, "n_jobs": -1}
tsne_kwargs["max_iter" if tuple(int(x) for x in sklearn.__version__.split(".")[:2]) >= (1, 4) else "n_iter"] = 1000

from diffusion_model.dataset import create_dataset
from diffusion_model.model import Stage1Model, Stage2Model, Stage3Model
from diffusion_model.model_loader import load_checkpoint

# ── CONFIG ───────────────────────────────────────────────────────────────────
STAGE1_CKPT  = "new_checkpoints/gcnn_stage1/stage1_best.pt"
STAGE2_CKPT  = "new_checkpoints/stage1_gcnn_mseonly/stage2_best.pt"
STAGE3_CKPT  = "new_checkpoints/stage3_gcnn/stage3_best.pt"
SKEL_FOLDER  = "/home/qsw26/smartfall/gait_loss/filtered_skeleton_data"
HIP_FOLDER   = "/home/qsw26/smartfall/gait_loss/SmartFallMM-Dataset/young/accelerometer/phone"
WRIST_FOLDER = "/home/qsw26/smartfall/gait_loss/SmartFallMM-Dataset/young/accelerometer/watch"
ENCODER_TYPE = "gcn"
IMU_GRAPH    = "multiscale"
VAL_FRAC     = 0.2
BATCH_SIZE   = 16
DDIM_STEPS   = 20
MAX_S3_SAMPLES = 400      # Stage-3 subset (DDIM is slow); panels 1-3 use full dataset
PCA_COMPONENTS = 64
PROBE_EPOCHS   = 50
NUM_CLASSES    = 14
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
OUT_PATH = "outputs/tsne_4stage_comparison.png"

# ── Load models ───────────────────────────────────────────────────────────────
device = torch.device(DEVICE)
print("Loading Stage-1 …")
stage1 = Stage1Model(latent_dim=256, num_joints=32, timesteps=500,
                     gait_metrics_dim=9, encoder_type=ENCODER_TYPE).to(device)
load_checkpoint(STAGE1_CKPT, stage1, strict=False)
stage1.eval()
for p in stage1.parameters(): p.requires_grad_(False)

print("Loading Stage-2 …")
stage2 = Stage2Model(encoder=stage1.encoder, latent_dim=256, num_joints=32,
                     gait_metrics_dim=9, num_classes=NUM_CLASSES,
                     imu_graph_type=IMU_GRAPH).to(device)
load_checkpoint(STAGE2_CKPT, stage2, strict=False)
stage2.eval()
for p in stage2.parameters(): p.requires_grad_(False)

print("Loading Stage-3 …")
stage3 = Stage3Model(
    encoder=stage1.encoder, decoder=stage1.decoder, denoiser=stage1.denoiser,
    latent_dim=256, num_joints=32, num_classes=NUM_CLASSES,
    timesteps=500, gait_metrics_dim=9,
).to(device)
load_checkpoint(STAGE3_CKPT, stage3, strict=False)
stage3.eval()
for p in stage3.parameters(): p.requires_grad_(False)

# ── Dataset ───────────────────────────────────────────────────────────────────
print("Building dataset …")
ds = create_dataset(None,
    skeleton_folder=SKEL_FOLDER, hip_folder=HIP_FOLDER, wrist_folder=WRIST_FOLDER,
    window=90, joints=32, num_classes=NUM_CLASSES, stride=45,
    normalize_sensors=True, gait_cache_dir="outputs/gait_cache",
)
n_val = int(len(ds) * VAL_FRAC)
train_ds, val_ds = random_split(ds, [len(ds)-n_val, n_val],
                                generator=torch.Generator().manual_seed(42))
print(f"  train={len(train_ds)}  val={len(val_ds)}")

# ── Encode all 4 representations ─────────────────────────────────────────────
@torch.no_grad()
def encode_all(loader, collect_s3=False, max_s3=MAX_S3_SAMPLES):
    raw_list, skel_list, imu_list, y_list = [], [], [], []
    s3_list, y_s3_list = [], []   # ← separate label list for Stage-3 (subset)
    s3_count = 0

    for batch in loader:
        x            = batch["skeleton"].to(device)
        a_hip        = batch["A_hip"].to(device)
        a_wrist      = batch["A_wrist"].to(device)
        gait_metrics = batch["gait_metrics"].to(device)
        y_batch      = batch["label"].numpy()
        B = x.shape[0]

        # Panels 1-3: always collected (full dataset)
        raw_list.append(x.cpu().numpy().reshape(B, -1))
        z0_real = stage1.encoder(x, gait_metrics=None)
        skel_list.append(z0_real.cpu().numpy().reshape(B, -1))
        h_tokens, h_global = stage2.aligner(a_hip_stream=a_hip, a_wrist_stream=a_wrist)
        imu_list.append(h_tokens.cpu().numpy().reshape(B, -1))
        y_list.append(y_batch)

        # Panel 4: Stage-3 DDIM — capped at max_s3, track its own labels
        if collect_s3 and s3_count < max_s3:
            z0_gen = stage3.diffusion.p_sample_loop_ddim(
                denoiser=stage3.denoiser,
                shape=torch.Size([B, x.shape[1], x.shape[2], 256]),
                device=device,
                h_tokens=h_tokens,
                h_global=h_global,
                gait_metrics=gait_metrics,
                sample_steps=DDIM_STEPS,
            )
            s3_list.append(z0_gen.cpu().numpy().reshape(B, -1))
            y_s3_list.append(y_batch)   # ← labels only for this subset
            s3_count += B

    return (
        np.concatenate(raw_list),
        np.concatenate(skel_list),
        np.concatenate(imu_list),
        np.concatenate(y_list),
        np.concatenate(s3_list)   if s3_list   else None,
        np.concatenate(y_s3_list) if y_s3_list else None,
    )

print("Encoding train …")
raw_tr, skel_tr, imu_tr, y_tr, s3_tr, y_s3_tr = encode_all(
    DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=0), collect_s3=True)
print("Encoding val …")
raw_va, skel_va, imu_va, y_va, s3_va, y_s3_va = encode_all(
    DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=0), collect_s3=True)
print(f"  Stage-3 train samples: {len(s3_tr)}  val samples: {len(s3_va)}")

# ── PCA ───────────────────────────────────────────────────────────────────────
def pca_compress(tr, va):
    n = min(PCA_COMPONENTS, tr.shape[0], tr.shape[1])
    pca = PCA(n_components=n, random_state=42)
    return pca.fit_transform(tr), pca.transform(va)

print("PCA …")
raw_tr_p,  raw_va_p  = pca_compress(raw_tr,  raw_va)
skel_tr_p, skel_va_p = pca_compress(skel_tr, skel_va)
imu_tr_p,  imu_va_p  = pca_compress(imu_tr,  imu_va)
s3_tr_p,   s3_va_p   = pca_compress(s3_tr,   s3_va)

# ── Linear probes ─────────────────────────────────────────────────────────────
def run_probe(tr, y_tr, va, y_va, name):
    probe = nn.Linear(tr.shape[1], NUM_CLASSES).to(device)
    opt   = torch.optim.Adam(probe.parameters(), lr=1e-3, weight_decay=1e-4)
    xt = torch.tensor(tr, dtype=torch.float32).to(device)
    yt = torch.tensor(y_tr, dtype=torch.long).to(device)
    xv = torch.tensor(va, dtype=torch.float32).to(device)
    yv = torch.tensor(y_va, dtype=torch.long).to(device)
    for _ in range(PROBE_EPOCHS):
        probe.train(); opt.zero_grad()
        nn.CrossEntropyLoss()(probe(xt), yt).backward(); opt.step()
    with torch.no_grad():
        acc = (probe(xv).argmax(1) == yv).float().mean().item()
    print(f"  [{name}] probe acc = {acc:.3f}")
    return acc

print("Linear probes …")
acc_skel = run_probe(skel_tr_p, y_tr,    skel_va_p, y_va,    "Stage-1 real z0")
acc_imu  = run_probe(imu_tr_p,  y_tr,    imu_va_p,  y_va,    "Stage-2 IMU")
acc_s3   = run_probe(s3_tr_p,   y_s3_tr, s3_va_p,   y_s3_va, "Stage-3 generated")  # ← own labels

# ── t-SNE ─────────────────────────────────────────────────────────────────────
# ── t-SNE ─────────────────────────────────────────────────────────────────────
print("t-SNE (takes ~5 min) …")
tsne = TSNE(n_components=2, random_state=42, perplexity=30, max_iter=1000)
emb_raw  = tsne.fit_transform(np.concatenate([raw_tr_p,  raw_va_p]));  print("  raw  done")
tsne = TSNE(n_components=2, random_state=42, perplexity=30, max_iter=1000)
emb_skel = tsne.fit_transform(np.concatenate([skel_tr_p, skel_va_p])); print("  skel done")
tsne = TSNE(n_components=2, random_state=42, perplexity=30, max_iter=1000)
emb_imu  = tsne.fit_transform(np.concatenate([imu_tr_p,  imu_va_p]));  print("  imu  done")
tsne = TSNE(n_components=2, random_state=42, perplexity=30, max_iter=1000)
emb_s3   = tsne.fit_transform(np.concatenate([s3_tr_p,   s3_va_p]));   print("  s3   done")


# ── Plot ──────────────────────────────────────────────────────────────────────
cmap = plt.get_cmap("tab20", NUM_CLASSES)
fig, axes = plt.subplots(1, 4, figsize=(30, 7))

def scatter_panel(ax, emb, labels, title, subtitle):
    for c in range(NUM_CLASSES):
        idx = labels == c
        if idx.sum():
            ax.scatter(emb[idx,0], emb[idx,1], s=12, alpha=0.65,
                       color=cmap(c), label=f"A{c+1}")
    ax.legend(markerscale=2, fontsize=6, ncol=2, loc="best")
    ax.set_title(f"{title}\n{subtitle}", fontsize=9)
    ax.set_xlabel("t-SNE-1"); ax.set_ylabel("t-SNE-2")

scatter_panel(axes[0], emb_raw,  all_y,    "Raw skeleton",              "Full sequence (PCA→t-SNE)")
scatter_panel(axes[1], emb_skel, all_y,    "Stage-1 skeleton latents",  f"Real z0  |  probe acc={acc_skel:.3f}")
scatter_panel(axes[2], emb_imu,  all_y,    "Stage-2 IMU latents",       f"h_tokens  |  probe acc={acc_imu:.3f}")
scatter_panel(axes[3], emb_s3,   all_y_s3, "Stage-3 generated latents", f"DDIM z0_gen ({DDIM_STEPS} steps)  |  probe acc={acc_s3:.3f}")

chance = 1.0 / NUM_CLASSES
fig.suptitle(
    f"All-stage representation comparison  (chance={chance:.3f})\n"
    f"Stage-1={acc_skel:.3f}  |  Stage-2 IMU={acc_imu:.3f}  |  Stage-3 generated={acc_s3:.3f}",
    fontsize=12, fontweight="bold",
)
plt.tight_layout()
os.makedirs("outputs", exist_ok=True)
plt.savefig(OUT_PATH, dpi=120, bbox_inches="tight")
plt.show()
print(f"\nSaved → {OUT_PATH}")
```

```python
# ── 4-Panel t-SNE (no PCA): Raw | Stage-1 | Stage-2 IMU | Stage-3 Generated ──
import sys, os
sys.path.insert(0, "/home/qsw26/smartfall/gait_loss/ldm_with_loss")

import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from torch.utils.data import DataLoader, random_split

from diffusion_model.dataset import create_dataset
from diffusion_model.model import Stage1Model, Stage2Model, Stage3Model
from diffusion_model.model_loader import load_checkpoint

# ── CONFIG ───────────────────────────────────────────────────────────────────
STAGE1_CKPT  = "new_checkpoints/gcnn_stage1/stage1_best.pt"
STAGE2_CKPT  = "new_checkpoints/stage1_gcnn_mseonly/stage2_best.pt"
STAGE3_CKPT  = "new_checkpoints/stage3_gcnn/stage3_best.pt"
SKEL_FOLDER  = "/home/qsw26/smartfall/gait_loss/filtered_skeleton_data"
HIP_FOLDER   = "/home/qsw26/smartfall/gait_loss/SmartFallMM-Dataset/young/accelerometer/phone"
WRIST_FOLDER = "/home/qsw26/smartfall/gait_loss/SmartFallMM-Dataset/young/accelerometer/watch"
ENCODER_TYPE = "gcn"
IMU_GRAPH    = "multiscale"
VAL_FRAC     = 0.2
BATCH_SIZE   = 16
DDIM_STEPS   = 20
MAX_S3_SAMPLES = 400
PROBE_EPOCHS   = 50
NUM_CLASSES    = 14
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
OUT_PATH = "outputs/tsne_direct_4stage_comparison.png"

# ── Load models ───────────────────────────────────────────────────────────────
device = torch.device(DEVICE)
print("Loading Stage-1 …")
stage1 = Stage1Model(latent_dim=256, num_joints=32, timesteps=500,
                     gait_metrics_dim=9, encoder_type=ENCODER_TYPE).to(device)
load_checkpoint(STAGE1_CKPT, stage1, strict=False)
stage1.eval()
for p in stage1.parameters(): p.requires_grad_(False)

print("Loading Stage-2 …")
stage2 = Stage2Model(encoder=stage1.encoder, latent_dim=256, num_joints=32,
                     gait_metrics_dim=9, num_classes=NUM_CLASSES,
                     imu_graph_type=IMU_GRAPH).to(device)
load_checkpoint(STAGE2_CKPT, stage2, strict=False)
stage2.eval()
for p in stage2.parameters(): p.requires_grad_(False)

print("Loading Stage-3 …")
stage3 = Stage3Model(
    encoder=stage1.encoder, decoder=stage1.decoder, denoiser=stage1.denoiser,
    latent_dim=256, num_joints=32, num_classes=NUM_CLASSES,
    timesteps=500, gait_metrics_dim=9,
).to(device)
load_checkpoint(STAGE3_CKPT, stage3, strict=False)
stage3.eval()
for p in stage3.parameters(): p.requires_grad_(False)

# ── Dataset ───────────────────────────────────────────────────────────────────
print("Building dataset …")
ds = create_dataset(None,
    skeleton_folder=SKEL_FOLDER, hip_folder=HIP_FOLDER, wrist_folder=WRIST_FOLDER,
    window=90, joints=32, num_classes=NUM_CLASSES, stride=45,
    normalize_sensors=True, gait_cache_dir="outputs/gait_cache",
)
n_val = int(len(ds) * VAL_FRAC)
train_ds, val_ds = random_split(ds, [len(ds)-n_val, n_val],
                                generator=torch.Generator().manual_seed(42))
print(f"  train={len(train_ds)}  val={len(val_ds)}")

# ── Encode all 4 representations ─────────────────────────────────────────────
@torch.no_grad()
def encode_all(loader, collect_s3=False, max_s3=MAX_S3_SAMPLES):
    raw_list, skel_list, imu_list, y_list = [], [], [], []
    s3_list, y_s3_list = [], []
    s3_count = 0

    for batch in loader:
        x            = batch["skeleton"].to(device)
        a_hip        = batch["A_hip"].to(device)
        a_wrist      = batch["A_wrist"].to(device)
        gait_metrics = batch["gait_metrics"].to(device)
        y_batch      = batch["label"].numpy()
        B = x.shape[0]

        raw_list.append(x.cpu().numpy().reshape(B, -1))
        z0_real = stage1.encoder(x, gait_metrics=None)
        skel_list.append(z0_real.cpu().numpy().reshape(B, -1))
        h_tokens, h_global = stage2.aligner(a_hip_stream=a_hip, a_wrist_stream=a_wrist)
        imu_list.append(h_tokens.cpu().numpy().reshape(B, -1))
        y_list.append(y_batch)

        if collect_s3 and s3_count < max_s3:
            z0_gen = stage3.diffusion.p_sample_loop_ddim(
                denoiser=stage3.denoiser,
                shape=torch.Size([B, x.shape[1], x.shape[2], 256]),
                device=device,
                h_tokens=h_tokens, h_global=h_global,
                gait_metrics=gait_metrics, sample_steps=DDIM_STEPS,
            )
            s3_list.append(z0_gen.cpu().numpy().reshape(B, -1))
            y_s3_list.append(y_batch)
            s3_count += B

    return (
        np.concatenate(raw_list),
        np.concatenate(skel_list),
        np.concatenate(imu_list),
        np.concatenate(y_list),
        np.concatenate(s3_list)   if s3_list   else None,
        np.concatenate(y_s3_list) if y_s3_list else None,
    )

print("Encoding train …")
raw_tr, skel_tr, imu_tr, y_tr, s3_tr, y_s3_tr = encode_all(
    DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=0), collect_s3=True)
print("Encoding val …")
raw_va, skel_va, imu_va, y_va, s3_va, y_s3_va = encode_all(
    DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=0), collect_s3=True)
print(f"  Stage-3 train samples: {len(s3_tr)}  val samples: {len(s3_va)}")
print(f"  Feature dims — raw:{raw_tr.shape[1]}  skel:{skel_tr.shape[1]}  imu:{imu_tr.shape[1]}  s3:{s3_tr.shape[1]}")

# ── Linear probes (on raw features, no PCA) ───────────────────────────────────
def run_probe(tr, y_tr, va, y_va, name):
    probe = nn.Linear(tr.shape[1], NUM_CLASSES).to(device)
    opt   = torch.optim.Adam(probe.parameters(), lr=1e-3, weight_decay=1e-4)
    xt = torch.tensor(tr, dtype=torch.float32).to(device)
    yt = torch.tensor(y_tr, dtype=torch.long).to(device)
    xv = torch.tensor(va, dtype=torch.float32).to(device)
    yv = torch.tensor(y_va, dtype=torch.long).to(device)
    for _ in range(PROBE_EPOCHS):
        probe.train(); opt.zero_grad()
        nn.CrossEntropyLoss()(probe(xt), yt).backward(); opt.step()
    with torch.no_grad():
        acc = (probe(xv).argmax(1) == yv).float().mean().item()
    print(f"  [{name}] probe acc = {acc:.3f}")
    return acc

print("Linear probes …")
acc_skel = run_probe(skel_tr, y_tr,    skel_va, y_va,    "Stage-1 real z0")
acc_imu  = run_probe(imu_tr,  y_tr,    imu_va,  y_va,    "Stage-2 IMU")
acc_s3   = run_probe(s3_tr,   y_s3_tr, s3_va,   y_s3_va, "Stage-3 generated")

# ── t-SNE (direct, no PCA) ────────────────────────────────────────────────────
# Warning: skel/s3 dims are ~700k+ — this will be slow and memory-heavy.
# If it OOMs, reduce MAX_S3_SAMPLES or add a sample cap below.
all_y    = np.concatenate([y_tr,    y_va])
all_y_s3 = np.concatenate([y_s3_tr, y_s3_va])

print("t-SNE (direct, no PCA — may take 10+ min) …")
tsne = TSNE(n_components=2, random_state=42, perplexity=30, max_iter=1000)
emb_raw  = tsne.fit_transform(np.concatenate([raw_tr,  raw_va]));  print("  raw  done")
tsne = TSNE(n_components=2, random_state=42, perplexity=30, max_iter=1000)
emb_skel = tsne.fit_transform(np.concatenate([skel_tr, skel_va])); print("  skel done")
tsne = TSNE(n_components=2, random_state=42, perplexity=30, max_iter=1000)
emb_imu  = tsne.fit_transform(np.concatenate([imu_tr,  imu_va]));  print("  imu  done")
tsne = TSNE(n_components=2, random_state=42, perplexity=30, max_iter=1000)
emb_s3   = tsne.fit_transform(np.concatenate([s3_tr,   s3_va]));   print("  s3   done")

# ── Plot ──────────────────────────────────────────────────────────────────────
cmap = plt.get_cmap("tab20", NUM_CLASSES)
fig, axes = plt.subplots(1, 4, figsize=(30, 7))

def scatter_panel(ax, emb, labels, title, subtitle):
    for c in range(NUM_CLASSES):
        idx = labels == c
        if idx.sum():
            ax.scatter(emb[idx,0], emb[idx,1], s=12, alpha=0.65,
                       color=cmap(c), label=f"A{c+1}")
    ax.legend(markerscale=2, fontsize=6, ncol=2, loc="best")
    ax.set_title(f"{title}\n{subtitle}", fontsize=9)
    ax.set_xlabel("t-SNE-1"); ax.set_ylabel("t-SNE-2")

scatter_panel(axes[0], emb_raw,  all_y,    "Raw skeleton",              "Direct t-SNE")
scatter_panel(axes[1], emb_skel, all_y,    "Stage-1 skeleton latents",  f"Real z0  |  probe acc={acc_skel:.3f}")
scatter_panel(axes[2], emb_imu,  all_y,    "Stage-2 IMU latents",       f"h_tokens  |  probe acc={acc_imu:.3f}")
scatter_panel(axes[3], emb_s3,   all_y_s3, "Stage-3 generated latents", f"DDIM z0_gen ({DDIM_STEPS} steps)  |  probe acc={acc_s3:.3f}")

chance = 1.0 / NUM_CLASSES
fig.suptitle(
    f"All-stage representation comparison — Direct t-SNE  (chance={chance:.3f})\n"
    f"Stage-1={acc_skel:.3f}  |  Stage-2 IMU={acc_imu:.3f}  |  Stage-3 generated={acc_s3:.3f}",
    fontsize=12, fontweight="bold",
)
plt.tight_layout()
os.makedirs("outputs", exist_ok=True)
plt.savefig(OUT_PATH, dpi=120, bbox_inches="tight")
plt.show()
print(f"\nSaved → {OUT_PATH}")
```
